```python
import pandas as pd
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt
conexion = sqlite3.connect("torneo_smash.db")

consulta_sql = """
SELECT Nombre, Personaje, Edad, Sexo
FROM combates
ORDER BY Edad DESC
"""
df_filtrado = pd.read_sql_query(consulta_sql, conexion)
print("📊 Datos extraídos mediante SQL con éxito:")
df_filtrado.head(10) # head(10) nos muestra los mejores 10 registros
```

```python
sns.set_theme(style="darkgrid")

plt.figure(figsize=(12, 6))

grafico_barras = sns.barplot(
    data=df_filtrado,
    x="Nombre",
    y="Edad",
    palette="magma",
    errorbar=None
)

plt.title("Edad absoluta de participantes", fontsize=16, fontweight="bold", pad=15)
plt.xlabel("Nombre de participantes", fontsize=12, labelpad=10)
plt.ylabel("Edad de participantes", fontsize=12, labelpad=10)


plt.xticks(rotation=45)


plt.show()
```
